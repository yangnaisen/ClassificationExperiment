# 进行数据可视化

seaborn是斯坦福大学开发的一个python数据可视化代码库，适合做复杂类型的数据可视化，基于matplotlib开发。代码的主页为：
http://seaborn.pydata.org/index.html

```python
%matplotlib inline
```

```python
import pandas as pd
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
```

如果没有安装seaborn，使用以下命令安装：

`conda install seaborn`

```python
#加载数据集Iris flower dataset，从网络下载
iris_source = "http://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
iris = pd.read_csv(iris_source, header=None, 
                   names=["SepalLengthCm","SepalWidthCm","PetalLengthCm","PetalWidthCm","Species"])
```

```python
#Seaborn库本身是自带iris这个数据集的
#iris=sns.load_dataset("iris")
```

```python
#输出前五个数据
iris.head()
```

```python
# 总览数据，看一下每个类别都包含多少个数据
iris["Species"].value_counts()
```

```python
#改变绘图的默认显示方式
mpl.style.use('ggplot')

#以下命令可以查看都存在哪些图表的显示方式
#mpl.style.available
```

```python
#选取前两个特征，绘制散点图进行可视化
iris.plot(kind="scatter", x="SepalLengthCm", y="SepalWidthCm")
```

```python
#seaborn的jointplot可以同时显示双变量的散点图和单变量的直方图
sns.jointplot(x="SepalLengthCm", y="SepalWidthCm", data=iris, size=5)
#size控制图形显示的大小
```

```python
# 上面的图像是没有每个样本的类别信息的，可以根据类别赋予不同的颜色
sns.FacetGrid(iris, hue="Species", size=5) \
   .map(plt.scatter, "SepalLengthCm", "SepalWidthCm") \
   .add_legend()
```

```python
#可以通过箱形图来查看单个特征
sns.boxplot(x="Species", y="PetalLengthCm", data=iris)
```

```python
#还可以更直观的地把原来的点的分布画到箱图上面
# 给这些点加上一点抖动 jitter=True，使得所有的点不是呈一条直线聚集在一起
#
#把每一个图的结果保存在同一个变量ax中，可以使得图形叠加
ax = sns.boxplot(x="Species", y="PetalLengthCm", data=iris)
ax = sns.stripplot(x="Species", y="PetalLengthCm", data=iris, jitter=True, edgecolor="gray")
```

```python
#把前面的两种图形的有点结合一下，就得到了小提琴图 violin plot 
#箱线图展示了分位数的位置，小提琴图则展示了任意位置的密度，
#通过小提琴图可以知道哪些位置的密度较高
sns.violinplot(x="Species", y="PetalLengthCm", data=iris, size=6)
```

```python
# Kernel density estimaton核密度估计替代直方图
sns.FacetGrid(iris, hue="Species", size=6) \
   .map(sns.kdeplot, "PetalLengthCm") \
   .add_legend()
```

```python
# pairplot 可以绘制出每两个特征之间的关系图
# 从特征两两组合的图中，可以看出 Iris-setosa 这个种类的花（红色）可以被任何一种的特征组合分离出来
sns.pairplot(iris, hue="Species", size=3)
#对角线上的直方图是单独一种特征下的频率分布
```

```python
# 也可以把对角线上的直方图替换成核密度估计的图
sns.pairplot(iris, hue="Species", size=3, diag_kind="kde")
```

```python
#也可以为每一个特征画出一个箱图
iris.boxplot(by="Species", figsize=(12, 6))
```

```python
# Andrews曲线将每个样本的属性值转化为傅里叶序列的系数来创建曲线
#通过将每一类曲线标成不同颜色可以可视化聚类数据，属于相同类别的样本的曲线通常更加接近
#注意坐标轴的范围。
from pandas.tools.plotting import andrews_curves
andrews_curves(iris, "Species")
```

```python
#使用平行坐标时，每个点用线段联接。
#每个垂直的线代表一个属性。一组联接的线段表示一个数据点，可能是一类的数据点会更加接近。
#坐标轴的顺序会影响可视化的效果
from pandas.tools.plotting import parallel_coordinates
parallel_coordinates(iris, "Species")
```

```python
# RadViz是基于弹簧压力最小化算法
# 简单来说，将一组点放在一个平面上，属性的原点放在一个单位圆上，
# 接下来你可以设想每个数据集通过一个弹簧联接到每个点上，弹力和他们属性值成正比（属性值已经标准化），
# 数据集在平面上的位置是弹簧的均衡位置，不同类的样本用不同颜色表示。
# 坐标轴的数值经过了标准化，三个特征的时候可视化效果最好
from pandas.tools.plotting import radviz
radviz(iris, "Species")
```
